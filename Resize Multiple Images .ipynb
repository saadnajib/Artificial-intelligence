```python
image_list=[]
resized_images=[]
```

```python
for filename in (glob.glob(r'C:/Users/saad najib/Desktop/saad/p/*.jpg')):
    print(filename)
    img = Image.open(filename)
    image_list.append(img)
#image_list = image_list[:10]
    
```

```python
for image in image_list:
    image = image.resize((300,300))
    resized_images.append(image)
```

```python
for (i,new) in enumerate(resized_images):
    new.save('{}{}{}'.format('C:/Users/saad najib/Desktop/p/saad',i+1,'.png'))
```
