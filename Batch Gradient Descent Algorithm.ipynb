```python
import random
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

#creating a dataset
price_y=[]
size_kanal_x=[]
for x in range(0,10):
    size_kanal_x.append(random.randint(1,3))
    if(size_kanal_x[x]==1):
        price_y.append(random.randint(100,900))
    elif(size_kanal_x[x]==2):
        price_y.append(random.randint(1000,2000))
    elif(size_kanal_x[x]==3):
        price_y.append(random.randint(2000,4000))

print(price_y)
print(size_kanal_x)
plt.plot(size_kanal_x,price_y, 'bo')
```

```python
theta0=0
theta1=0

#hypothesis
def hypo(x):
    return(theta0+theta1*x)
```

```python
def plot_line(data_points):
    x_values=[i for i in range( int(min(data_points))-1,int(max(data_points))+2)]
    y_values=[hypo(x) for x in x_values]
    plt.plot(x_values,y_values, 'r')
```

```python
plot_line(size_kanal_x)
```
